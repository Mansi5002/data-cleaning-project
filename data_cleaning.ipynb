```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("people-1000.csv")

print("Original Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
print("\nMissing Values:\n")
print(df.isna().sum())
```

```python
print("\nMissing Values:\n")
print(df.isna().sum())
```

```python
print("\nDuplicate Records:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print("Duplicate Records After Removal:", df.duplicated().sum())
```

```python
if "Index" in df.columns:
    df = df.drop(columns=["Index"])

df.head()
```

```python
df["Date of birth"] = pd.to_datetime(df["Date of birth"], format="%d-%m-%Y")
```

```python
df["Phone"] = df["Phone"].astype(str)
df["Phone"] = df["Phone"].str.replace(r"\D", "", regex=True)
```

```python
text_columns = ["First Name", "Last Name", "Sex", "Email", "Job Title"]

for col in text_columns:
    df[col] = df[col].str.lower().str.strip()

df.head()
```

```python
print("Final Shape:", df.shape)
print("Remaining Duplicates:", df.duplicated().sum())
print("\nRemaining Missing Values:\n", df.isna().sum())

assert df.duplicated().sum() == 0
```

```python
df.to_csv("cleaned_people_1000.csv", index=False)

print("Cleaned dataset saved successfully.")
```
